```python
# подсчитываем количество недель в каждом месяце
orders_by_month['count_of_weeks']=(orders_by_month['count_of_days']/7).round(2)
```

```python
# получаем среднее количество заказов в неделю по месяцам по каждому пользователю
orders_by_month['avg_orders_by_week']=(orders_by_month['count_orders']/orders_by_month['count_of_weeks']).round(3)
```

# 6) Провести RFM-сегментацию пользователей

В кластеризации выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты, 
F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. 
Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary 
для интерпретации этих кластеров.

```python
# отбираем пользователей, у которых покупка прошла "успешно"
customers_orders=customers_orders[customers_orders['order_approved_at'].notnull()].query("order_status !=('unavailable','canceled')")
```

Создаем датафрейм с информацией о стоимости заказов и выполняем join с основным датафреймом, в котором информация о пользователях и заказах

```python
price_orders=items_dataset.groupby('order_id', as_index=False).agg({'price':sum}).rename(columns={'price' : 'price_orders'})
```

```python
customers_orders=customers_orders.merge(price_orders, how='left', on='order_id')
```

Посмотрим когда был запрос на первый и последний заказ, и временной промежуток между ними

```python
print('Orders from {} to {}'.format(customers_orders['order_purchase_timestamp'].min(), customers_orders['order_purchase_timestamp'].max()))
```

```python
customers_orders['order_purchase_timestamp'].max() - customers_orders['order_purchase_timestamp'].min()
```

```python
# создаем переменную с датой следующей после последней покупки
now=customers_orders['order_purchase_timestamp'].max() + timedelta(days=1)
```

```python
# создаем колонку с данными о давности выполнения заказа
customers_orders['days_since_order'] = customers_orders['order_purchase_timestamp'].apply(lambda x: (now - x).days)
```

```python
# создаем rfm для каждого покупателя
rfm=customers_orders.groupby('customer_unique_id',as_index=False)\
    .agg({'price_orders':'sum', 'order_id':'count','days_since_order':'min' })\
    .rename(columns={'days_since_order' : 'recency', 'order_id' : 'frequency', 'price_orders' : 'monetary'})
```

```python
# визуализируем квантили
quintiles = rfm[['recency', 'frequency', 'monetary']].quantile([.2, .4, .6, .8]).to_dict()
quintiles
```

```python
# данные по квантилям(границам метрик) не детализируют параметр frequency, посмотрим подробнее про эту колонку
rfm.frequency.value_counts()
```

```python
# присваиваем границам метрик баллы
def r_score(x):
    if x <= quintiles['recency'][.2]:
        return 5
    elif x <= quintiles['recency'][.4]:
        return 4
    elif x <= quintiles['recency'][.6]:
        return 3
    elif x <= quintiles['recency'][.8]:
        return 2
    else:
        return 1

    
def f_score(x):
    if x == 1:
        return 1
    elif x ==2:
        return 2
    else:
        return 3
    
def m_score(x):
    if x <= quintiles['monetary'][.2]:
        return 1
    elif x <= quintiles['monetary'][.4]:
        return 2
    elif x <= quintiles['monetary'][.6]:
        return 3
    elif x <= quintiles['monetary'][.8]:
        return 4
    else:
        return 5
```

```python
rfm['R'] = rfm['recency'].apply(lambda x: r_score(x))
rfm['F'] = rfm['frequency'].apply(lambda x: f_score(x))
rfm['M'] = rfm['monetary'].apply(lambda x: m_score(x))
```

```python
# создаем новый столбец с последовательностью баллов
rfm['rfm_score'] = rfm['R'].map(str) + rfm['F'].map(str) + rfm['M'].map(str)
```

```python
# описываем сегменты относительно 2 параметров: recency и frequency
segt_map = {
    r'[1-2]1': 'inactive',
    r'[1-2]2': 'at risk',
    r'[1-2]3': 'can not loose',
    r'31': 'doubt',
    r'32': 'need attention',
    r'[3-4]3': 'loyal customers',
    r'41': 'hopeful',
    r'51': 'new customers',
    r'[4-5]2': 'potential loyalists',
    r'53': 'ideal customer'
}
rfm['segment'] = rfm['R'].map(str) + rfm['F'].map(str)
rfm['segment'] = rfm['segment'].replace(segt_map, regex=True)
```

```python
rfm.head()
```

```python
# визуализируем соотношение сегментов
segments_counts = rfm['segment'].value_counts().sort_values()
fig, ax = plt.subplots()
bars = ax.barh(range(len(segments_counts)),
              segments_counts,
              color='silver')
ax.set_frame_on(False)
ax.tick_params(left=False,
               bottom=False,
               labelbottom=False)
ax.set_yticks(range(len(segments_counts)))
ax.set_yticklabels(segments_counts.index)

for i, bar in enumerate(bars):
        value = bar.get_width()
        if segments_counts.index[i] in ['ideal customer', 'loyal customers','potential loyalists','can not loose']:
            bar.set_color('firebrick')
        ax.text(value,
                bar.get_y() + bar.get_height()/2,
                '{:,} ({:}%)'.format(int(value),
                                   int(value*100/segments_counts.sum())),
                va='center',
                ha='left'
               )
plt.show()  
```
